```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img_path = "Your Image Path"

img = cv2.imread(img_path)
mean = 0
var = 10
sigma = var ** 0.5
gaussian = np.zeros((368, 448), np.float32)

noisy_image = np.zeros(img.shape, np.float32)  #np.random.normal(mean, sigma, (224, 224))

if len(img.shape) == 2:
    noisy_image = img + gaussian
else:
    noisy_image[:, :, 0] = img[:, :, 0] + gaussian
    noisy_image[:, :, 1] = img[:, :, 1] + gaussian
    noisy_image[:, :, 2] = img[:, :, 2] + gaussian

cv2.normalize(noisy_image, noisy_image, 0, 255, cv2.NORM_MINMAX, dtype=-1)
noisy_image = noisy_image.astype(np.uint8)
Ii = np.random.randn(368,448)*10 + noisy_image.mean(2)  #imaginary channel
Ir = np.random.randn(368,448)*10 + noisy_image.mean(2)  #real channel
noisyimage2 = (Ii**2 + Ir**2)**0.5

#cv2.imshow("img", img)
#cv2.imshow("gaussian", gaussian)

plt.imshow(noisyimage2,cmap='gray')
plt.show()

cv2.waitKey(0)
```
